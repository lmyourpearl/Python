```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams["font.size"] =12
plt.rcParams['xtick.labelsize'] =12.
plt.rcParams['ytick.labelsize'] =12.
plt.rcParams['axes.unicode_minus'] = False

CHART_NAME = 'humanTest'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = './data/human_info.csv'

import pandas as pd
humaninfo = pd.read_csv(filename, encoding='cp949')
```

```python
humaninfo.head(10)
```

```python
humaninfo.describe()
```

```python
# filesave 함수
def save():
    global cnt
    cnt += 1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi=400)
    print(savefile + '파일이 저장되었습니다.')
# end
    
```

```python
import seaborn as sns

ORDERING = ['결혼', '이혼', '미혼']
plt.figure()
plt.title('결혼 유무에 따른 빈도')
sns.countplot(
    x='marriage',
    data=humaninfo,
    order=ORDERING,
    hue='marriage',
    palette='Set2',
    legend=False)
save()
```

```python
plt.figure()
plt.title('결혼 유무 VS 종교 유무 빈도')
sns.countplot(
    x='marriage',
    hue='religion',
    order=ORDERING,
    data=humaninfo
)
save()
```

```python
x = humaninfo['age']

plt.figure()
plt.title('kde와 histogram')
sns.histplot(x, kde=True, label='kdehisto', edgecolor='none')
save()
```

```python
pivot = humaninfo.pivot_table(index='gender', columns='marriage', values='age')
pivot
```

```python
plt.figure()
plt.title('결혼 유무와 성별에 대한 히트맵')
sns.heatmap(data = pivot, annot=True)
save()
```

```python
cor = humaninfo.corr(numeric_only=True)
cor
```

```python
plt.figure()
plt.title('상관 관계 시각화')
sns.heatmap(data=cor, annot=True, cmap='YlGnBu')
save()
```

```python
newhumaninfo = humaninfo.loc[:,['marriage','income', 'age']]

sns.pairplot(data=newhumaninfo, hue='marriage')
save()
```

```python
plt.figure()
plt.title('나이에 대한 바이올린 그래프')
sns.violinplot(x='age', data=humaninfo)
save()
```

```python
g= sns.lmplot(x='age', y='income', hue='marriage', height=8, data=humaninfo)
g.fig.suptitle('나이와 소득에 대하여 결혼 유무별 lmplot 그래프', y=1.03, fontsize=15)
save()
```

```python
sns.relplot(x='age', y='income',hue='gender', col='marriage', data=humaninfo, palette='CMRmap_r')
save()
```

```python
sns.jointplot(x='age', y='income', height=8, data=humaninfo)
save()
```

```python
humaninfo[['income','age']].describe()
```

```python
ax = plt.subplots
ax = sns.barplot(x='income', y='gender', hue='religion', data=humaninfo, err_kws={'linewidth':0})
ax.set_title('소득과 성별에 대하여 종교 유무별 그래프')
ax.set_ylabel('gender')
ax.set_xlabel('income')
save()
```
