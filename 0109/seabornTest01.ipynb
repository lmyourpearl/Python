```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams["font.size"] = 12
plt.rcParams['xtick.labelsize'] = 12.
plt.rcParams['ytick.labelsize'] = 12.
plt.rcParams['axes.unicode_minus'] = False

CHART_NAME = 'seabornTest01'
cnt, PNG, UNDERBAR = 0, '.png', '_'
filename = './data/tips.csv'

tips = pd.read_csv(filename, encoding='utf-8')
```

```python
tips.head(10)
```

```python
tips.describe()
```

```python
def save():
    global cnt
    cnt += 1
    savefile = CHART_NAME + UNDERBAR + str(cnt).zfill(2) + PNG
    plt.savefig(savefile, dpi=400)
    print(savefile + '파일이 저장되었습니다.')
#end
```

```python
import seaborn as sns

total = tips['total_bill']
```

```python
import numpy as np

print(np.max(total.unique()))
print(np.min(total.unique()))
```

```python
fig, ax = plt.subplots()
sns.scatterplot(data=tips, x='total_bill', y='tip', ax=ax)
ax =sns.rugplot(data=tips, x='total_bill', y='tip', ax=ax)
ax.set_title('산점도와 Rug Plot')
ax.set_xlabel('총 지급액')
ax.set_ylabel('팁(tip)')
save()
```

```python
fig, ax = plt.subplots()
sns.histplot(data=tips, x='total_bill', kde=True, bins=30, ax=ax)
ax.set_title('총 지급액 히스토그램')
ax.set_xlabel('총 지급액')
ax.set_ylabel('')
save()
```

```python
plt.figure()
ax = sns.barplot(x='day', y='tip', data=tips, hue='time', errorbar=None)
ax.set_title('요일에 따른 팁(tip)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
ax.set_xticks(range(len(['일요일', '토요일', '목요일', '금요일'])))
ax.set_xticklabels(['일요일', '토요일', '목요일', '금요일'])
ax.legend(title='time', loc='lower left')
save()
```

```python
plt.figure()
outlier_style = dict(markerfacecolor='gray', marker='d')
ax = sns.boxplot(x='day', y='tip', hue='time', data=tips, flierprops=outlier_style)
ax.set_title('요일에 따른 팁(tip)(boxplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
ax.set_xticks(range(len(['일요일', '토요일', '목요일', '금요일'])))
ax.set_xticklabels(['일요일', '토요일', '목요일', '금요일'])
ax.legend(title='time', loc='upper right')
save()
```

```python
plt.figure()
ax = sns.violinplot(x='day', y='tip', hue='time', data=tips)
ax.set_title('요일에 따른 팁(tip)(violinplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
ax.set_xticks(range(len(['일요일', '토요일', '목요일', '금요일'])))
ax.set_xticklabels(['일요일', '토요일', '목요일', '금요일'])
ax.legend(title='time',loc='upper right')
save()
```

```python
plt.figure()
ax = sns.stripplot(x='day', y='tip', hue='time', data=tips)
ax.set_title('요일에 따른 팁(tip)(stripplot)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
ax.set_xticks(range(len(['일요일', '토요일', '목요일', '금요일'])))
ax.set_xticklabels(['일요일', '토요일', '목요일', '금요일'])
ax.legend(title='time',loc='upper right')
save()
```

```python
plt.figure()
ax = sns.swarmplot(x='day', y='tip', hue='time', data=tips)
ax.set_title('Swarm Plot(hue=time)')
ax.set_xlabel('요일')
ax.set_ylabel('팁(tip)')
ax.set_xticks(range(len(['일요일', '토요일', '목요일', '금요일'])))
ax.set_xticklabels(['일요일', '토요일', '목요일', '금요일'])
ax.legend(loc='upper right')
save()
```

```python
import numpy as np
plt.figure()
sns.boxplot(x='total_bill', y='day', hue='day', data=tips, whis=np.inf)
ax = sns.stripplot(x='total_bill', y='day', hue='day', data=tips, jitter=True, palette='dark:0.4')
ax.set_title('Boxplot고 Strip Plot로 표현')
ax.set_xlabel('총 지급액')
ax.set_ylabel('요일')
save()
```
