# 반복문(for)

#for(int i = 0; i < 11; i++){}

# for in range(시작값, 끝값, 증감값)

```python
for i in range(0, 10, 1):
  print(i)
```

```python
for i in range(0, 10):
  print(i)
# 증감값 생략 가능
```

# 증감값, 시작값 생략 
#시작값의 기본은 0, 증감은 1이 들어감 

```python
for i in range(10):
  print(i)
```

```python
for i in range(0, 10, 2):
  print(i)
```

```python
for i in 'abc':
  print(i)
```

# List

```python
for i in [1,2,3,]:
    print(i)
```

# for if 

```python
for i in [1,2,3]:
    if( i % 2 == 0):
        print(i)
```
