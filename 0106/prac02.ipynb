# 디폴트 매개변수

```python
# def 함수이름(매개변수1, 매개변수2 = A, 매개변수3 = B):
    # 수행 문장
    # return 반환값

# ** 뒤에서부터 선언해야 함. a,b,c가 있다면 c부터 
```

```python
def func1(a, b=5, c=10):
    return a + b + c

print(func1(1, 2, 3))
print(func1(1, 2))
print(func1(1))

# 1 (a) + 5 (b) + 10 (c) = 16
```

# 매개변수 N개

```python
# def 함수이름(*매개변수):
    # 수행문장
    # return 반환값

# ** args(argument) n개의 매개변수를 표현할 때 사용 
```

```python
def func6(*args):
    a = 0
    for i in args:
        a = a + i
    return a

b = func6(1,2)
print(b)

c = func6(2,3,4,5)
print(c)

d = func6(1,2,3,4,5,4,3,2,1)
print(d)

e = func6()
print(e)
```

```python
str1 = "aaaaa"
str2 = "bbbbb"
result = "".join(a + b for a, b in zip(str1, str2))
print(result)
```

```python
def solution(str1, str2):
    result = "".join(a + b for a,b in zip(str1, str2))
    return result

str1 = "aaaaa"
str2 = "bbbbb"
answer = solution(str1, str2)
print(answer)
```
