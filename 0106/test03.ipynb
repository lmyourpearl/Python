```python
# 성적 리스트를 입력 받아 평균을 출력하는 print_score 함수를 정의하라. print_score ([1, 2, 3])
# 성적 총 합, 성적의 개수 - sum(), len()
```

```python
def print_score(list):
    avg = sum(list) / len(list)
    print(avg)
    
print_score([1,2,3])
```

```python
# 하나의 리스트를 입력받아 짝수만 화면에 
#출력하는 print_even 함수를 정의하라. print_even ([1, 3, 2, 10, 12, 11, 15])

#나머지 연산자 %
```

```python
def print_even(list):
    for i in list:
        if i % 2 == 0:
            print(i)

print_even([1,3,2,10,12,11,15])
```

```python
def get_even_list(num_list):
    result = [n for n in num_list if n % 2 == 0]
    print(result)
get_even_list([1,3,2,10,12,11,15])
```
