# 딕셔너리 선언

```python
ice = {"메로나":1000, "폴라포":1200, "빵빠레":1800}
print(ice["메로나"])
```

# add

```python
ice["죠스바"] = 1200
ice["월드콘"] = 1500
print(ice)
```

```python
inventory = {"메로나": [300,20],
             "비비빅": [400,3],
             "죠스바": [250,100]
             }
print(inventory)
print(inventory["메로나"][0])
```

```python
# example
person = {
  "name": "Alice",
  "age": 30,
  "city": "New York",
  "job": "Software Engineer"
}
print(person["name"])

student_grades = {
  "John": 85,
  "Sarah": 90,
  "Anna": 78,
  "Tom": 92
}
johns_grade = student_grades.get("John")
print("John's grade:", johns_grade)
```
