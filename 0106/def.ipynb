# 선언하고 호출 **

```python
def greet():
  print("Hello, World!")
greet()
```

```python
def print_name(name):
  print(f"Hello, {name}!")
print_name("JJ")
```

```python
def get_pi():
  return 3.141592
pi_value = get_pi()
print(pi_value)
```

```python
# 매개변수가 있는 함수 선언
def add(a,b):
  print(a+b)
# 인수 있는 함수 호출
add(10,20)
```

```python
def addd(a,b):
  return a+b
sum_result = addd(3,5)
print(sum_result)
```

```python
def sum_all(*args):
  total = 0
  for num in args:
    total += num
  return total

result = sum_all(1,2,3,4,5)
print(result)

result2 = sum_all(10,20,30)
print(result2)
```

```python
# 매개변수, return이 있는 함수 선언
def plus(a, b):
  return a + b
# 함수 호출
result = plus(10,20)
print(result)
```
