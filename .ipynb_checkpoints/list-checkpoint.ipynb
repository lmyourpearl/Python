# 자바의 배열 같은 것
# list
# 배열 요소 값 수정, 추가 가능

```python
movieList=["닥터 스트레인지", "스플릿", "럭키"]
print(movieList)
print(movieList[1])
#add '배트멘'
movieList.append("배트맨")
print(movieList)
# edit '배트맨3'
movieList[3] = "배트맨3"
print(movieList)
del movieList[2]
print(movieList)
```

# tuple(튜플)
# 배열의 요소(값) 수정, 추가 불가능

```python
movieTuple=("닥터 스트레인지", "스플릿", "럭키")
print(movieTuple)
print(movieTuple[1])
# movieTuple.append("배트맨") - 추가 불가능
# moviceTuple[3] = "배트맨3" - 수정 불가능
## 값이 고정되어야하면 tuple에 저장, 변경되면 list에 저장 
```

# 리스트 값 합계, 평균, 최대값, 최소값

```python
score = [65,73,84,98,70]
print(sum(score))
print(max(score))
print(min(score))
print(sum(score)/len(score))
```
