# 데이터 병합하기

# merge()

```python
import pandas as pd

dict1 = {'name' :['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 'korean': range(7)}
df1 = pd.DataFrame(dict1)

print( f'\n# DataFrame 출력 01 \n{'-'*40}')
df1
```

```python
dict2 = {'name':['김철수', '홍길동', '심수봉'], 'english':range(3)}
df2 = pd.DataFrame(dict2)

print( f'\n# DataFrame 출력 02 \n{'-'*40}')
df2
```

```python
print( f'\n# merge() 메소드의 on="name"을 이용하여 데이터 합치기 \n{'-'*60}')
pd.merge(df1, df2, on='name')
```

```python
print( f'\n# 이름순으로 정렬 \n{'-'*60}')
pd.merge(df1, df2, on='name').sort_values(by='name')
```

```python
print(f'\n# how="outer" 라고 명시하면 full outer join이다. \n{'-'*50}')
pd.merge(df1, df2, how= 'outer')
```

```python
print( f'\n# 컬럼 이름이 동일하지 않는 경우 \n{'-'*40}')
dict3 = {'leftkey':['홍길동', '홍길동', '김철수', '박영희', '김철수', '김철수', '홍길동'], 'korean':range(7)}
df3 = pd.DataFrame(dict3)
print( '# DataFrame 출력 03')
df3
```

```python
dict4 = {'rightkey': ['김철수', '홍길동', '심수봉'], 'english':range(3)}
df4 = pd.DataFrame(dict4)
print( f'\n# DataFrame 출력 04\n{'-'*30}')
df4
```

```python
print( f'\n# merge() 메소드의 left_on과 right_on 사용하기 \n{'-'*50}')
pd.merge(df3, df4, left_on='leftkey', right_on='rightkey')
```

```python
dict1 = { 'key1':['김철수', '김철수', '박영희'],
          'key2':['one', 'two', 'one'],
          'leftval_x': [1,2,3]}
left = pd.DataFrame(dict1)
print( f'\n# DataFrame 출력 05 \n{'-'*30}')
left
```

```python
dict2 = {'key1':['김철수', '김철수', '박영희', '박영희'],
         'key2':['one','one','one','two'],
         'leftVal_y':[4,5,6,7]}
right = pd.DataFrame(dict2)
print( f'\n# DataFrame 출력 06 \n{'-'*30}')
right
```

```python
# 조인할 컬럼 리스트
mylist = ['key1', 'key2']
print( f'\n# 여러 개의 컬럼 병합하기 \n{'-'*30}')
pd.merge(left, right, on=mylist, how='outer')
```

```python
print( f'\n# suffixes 옵션 사용하기 \n{'-'*30}')
print( '# 동일한 컬럼 이름 key2에 대하여 접미사를 붙여 준다.')
pd.merge(left, right, on='key1', suffixes=('_왼쪽', '_오른쪽'))
```

```python
print( f'\n# 색인을 이용한 머지 사용하기 \n{'-'*30}')
newdf1 = df1.set_index('name')
newdf1
```

```python
newdf2 = df2.set_index('name')
newdf2
```

```python
print( f'\n# 색인을 이용한 병합 \n{'-'*30}')
pd.merge(newdf1, newdf2, left_index=True, right_index=True, how='outer', indicator = True)
```
