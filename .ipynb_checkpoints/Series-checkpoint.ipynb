```python
import pandas as pd
import numpy as np

sdata = [2000.0, 3000.0, 4000.0, np.nan]
city = ['서울', '부산', '울산', '목포']
myseries2 =  pd.Series(sdata, index=city)

#객체 자체의 이름
myseries2.name = '도시 데이터'
#색인의 이름
myseries2.index.name = '도시'
print ( myseries2 )
```

```python
# series

import pandas as pd

myindex = ['서울', '부산', '광주', '대구', '울산', '목포', '여수'] 
#변수(list 자료형 사용)

mylist = [50, 60, 40, 80, 70, 30, 20]
myseries = pd.Series(data=mylist, index=myindex)
print(myseries)
```
