# DataFrame

```python
# 데이터 분석용 라이브러리
import pandas as pd
# 수치 계산용 라이브러리
import numpy as np
```

```python
#행 인덱스
myindex = ['이순신', '김유신', '강감찬', '광해군', '연산군']

# 열 인덱스
mycolums = ['서울', '부산', '광주', '목포', '경주']

# 1~25까지의 숫자에 10을 곱한 값을 리스트로 저장
# range(1,26)
# onedata * 10
# list 함수 -> 위 결과를 list 형태로 저장하기 위한 함수
mylist = list( 10 * onedata for onedata in range(1,26))

# 출력
print(mylist)
```

```python
myframe = pd.DataFrame(np.reshape(mylist, (5,5)),
                       index = myindex,
                       columns = mycolums)
print(myframe)
```

# iloc

```python
from IPython.display import display

print(f"\n# 1 행만 읽어 오기\n{'-'*40}")
result = myframe.iloc[1]
print(type(result))
print(result)
# 시리즈를 데이터프레임으로 변환해서 표로 출력
#display(result.to_frame().T)
```

```python
# print(' \n# 몇개의 행을 읽어 오기 ')
print(f"\n# 몇개의 행을 읽어 오기 \n{'-'*40}")
result = myframe.iloc[[1,3]]
print(type(result))
#print(result)
result
```

```python
# print(' \n# 짝수행만 가져오기 ')
print(f"\n# 짝수행만 가져오기\n{'-'*40}")
result = myframe.iloc[0::2]
result
```

# loc

```python
print( '\n# 홀수행만 가져 오기')
print( f'\n# 이순신 행만 Series로 읽어오기 \n{'-'*40}')

result = myframe.loc['이순신']
print(type(result))
print(result)
```

```python
print( '\n# 홀수행만 가져 오기')
print( f'\n# 이순신 행만 DataFrame로 읽어오기 \n{'-'*40}')

result = myframe.loc[['이순신']]
result
```

```python
print(f'\n# 강감찬과 이순신 행 읽어 오기 \n{'-'*40}')
result = myframe.loc[['이순신', '강감찬']]
result
```

# np.random.choice() 함수: 복원 추출을 수행

```python
print(myframe.index)
```

```python
mytarget = np.random.choice(myframe.index, 3)
print(mytarget)
```

```python
result = myframe.loc[mytarget, mycolums]
result
```

```python
print( f'\n# 강감찬의 광주 실적 정보 가져오기 \n{'-'*40}')
result = myframe.loc[['강감찬'], ['광주']]
result
```

```python
print( f'\n# 연산군과 광해군의 광주/목포 정보 가져오기 \n{'-'*40}')
result = myframe.loc[['연산군', '강감찬'], ['광주', '목포']]
result
```

```python
print( f'\n# 연속적인 데이터 가져오기 \n{'-'*40}')
result = myframe.loc['김유신':'광해군', '광주':'목포']
result
```

```python
print( f'\n# 김유신~광해군까지 부산 실적 정보 가져오기  \n{'-'*40}')
result = myframe.loc['김유신' : '광해군', ['부산']]
result
```

```python
print( f'\n# Boolean으로 데이터 처리하기 \n{'-'*40}')
result = myframe.loc[[False, True, True, False, True]]
result
# true의 값만 보여줌
```

```python
print( f'\n# 부산 실적이 100이하인 항목들 \n{'-'*40}')
result = myframe.loc[myframe['부산'] <= 100]
result
```

```python
print( f'\n# 목포 실적이 140인 항목들 \n{'-'*40}')
result = myframe.loc[myframe['목포'] == 140]
result
```

```python
print( f'\n# 부산 실적이 70이상, 목포 실적이 140 이상인 항목 구하기 \n{'-'*40}')

cond1 = myframe['부산'] >= 70
cond2 = myframe['목포'] >= 140
print(type(cond1))
```

```python
print(cond1)
```

```python
print(cond2)
```

# all() 함수 = and

# any() 함수 = or 

```python
print( f'\n# 구한 두 개의 조건식을 합쳐서 데이터 프레임 df \n{'-'*40}')

df = pd.DataFrame([cond1, cond2])
display(df)

print(df.all())
```

```python
print(df.any())
```

```python
result = myframe.loc[df.all()]
result
```

```python
result = myframe.loc [df.any()]
result
```

```python
print( f'\n# 람다 함수의 사용 \n{'-'*40}')
result = myframe.loc[lambda df : df['광주'] >= 130]
result
```

```python
print( f'\n# 이순신과 강감찬의 부산 실적을 30으로 변경하기 \n{'-'*40}')
myframe.loc[['이순신', '강감찬'], ['부산']] = 30
myframe
```

```python
print( f'\n# 김유신부터 광해군까지 경주 실적을 80으로 변경하기 \n{'-'*40}')
myframe.loc['김유신' : '광해군', ['경주']] = 80
myframe
```

```python
print( f'\n# 연산군의 모든 실적을 50으로 변경하기 \n{'-'*40}')
myframe.loc[['연산군'],:] = 50
myframe
```

```python
print( f'\n# 모든 사람의 광주 컬럼을 60으로 변경하기 \n{'-'*40}')
myframe.loc[:, ['광주']] = 60
myframe
```

```python
print( f'\n# 경주 실적이 150이하인 데이터를 모두 0으로 변경하기 \n{'-'*40}')
myframe.loc[myframe['경주'] <= 150, ['경주', '광주']] = 0
myframe
```

```python
print( f'\n# 데이터 프레임 사용하기 \n{'-'*40}')
myframe
```
