# GroupBy 매카닉

# 데이터 분할(split) -> 반영(`) -> 결합(combine)

```python
df.head(n=10)
```

# unique() - 중복값을 제거하고, 유일값만 저장하는 numpy 객체 반환 = distinct

```python
df['marriage'].unique()
```

```python
print( f'\n# marriage컬럼에 대하여 그룹핑을 수행한 후 code_religion 컬럼의 평균 구하기 \n{'-'*80}')

df.groupby('marriage')['code_religion'].mean()
#결혼 상태(marriage)별로 사람들을 나누고, 각 그룹의 종교 유무(code_religion) 평균
# 기혼자 그룹과 미혼자 그룹 중 어느 쪽이 평균적으로 종교를 더 많이 가지고 있을까?
```

```python
mygrouping = df.groupby('marriage')
print(type(mygrouping))
print(mygrouping)
```

```python
grp_code_religion = mygrouping['code_religion']
print(type(grp_code_religion))
```

```python
mean_code_religion = grp_code_religion.mean()
print(mean_code_religion)
```

```python
print( f'\n# marriage, birth 컬럼을 이용하여 그룹화시킨 후 code_religion, incom에 대한 평균 값 출력 \n{'-'*80}')
mygrouping2 = df.groupby(['marriage', 'birth'])
result = mygrouping2[['code_religion', 'income']].mean()
result
```

```python
# 한줄 
df.groupby(['marriage', 'birth'])[['code_religion', 'income']].mean()
```
