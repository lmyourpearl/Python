# 함수 적용과 매핑(apply 함수)

# apply() - f = lambda x:x.max()-x.min()

# applymap(func) = format = lambda x: '%.2f 퍼센트'% x

```python
import pandas as pd

filename = './data/memberInfo.csv'
df = pd.read_csv(filename, index_col='id')

print(' 시리즈와 apply 메서드')
print(df)
```

```python
print('일반적인 산술 연산(각각 5점씩 더하기)')
print(df['kor'] + 5)
```

```python
# plus 함수 선언
def plus5(x):
    return x + 5
    
print( 'apply 함수를 적용한 결과')
#DataFrame에 내장함수인 apply()를 통해서 사용자정의 함수 plus를 호출
sq = df['kor'].apply(plus5)
print(sq)
```

```python
print( f'\n# 어떤 수의 n배를 곱해주는 함수 gob 정의 ')

def gob(x,n):
    return n * x
```

```python
print( f'\n# 국어(df.kor) 점수에 대하여 2배수의 값을 반환 \n{'-'*50}')

ex = df['kor'].apply(gob, n=2)
print(ex)
```

```python
print( f'\n# 국어(df.kor) 점수에 대하여 3배수의 값을 반환 \n{'-'*50}')
ex = df['kor'].apply(gob,n=3)
print(ex)
```

```python
print( f'\n# 컬럼 방향으로 평균 값을 구해주는 함수 applyByColumn(col) 구현 \n{'-'*70}')

def applyByColumn(col):
    mysum = 0
    for item in col:
        mysum += item
    return mysum / df.shape[0]

print(df.apply(applyByColumn, axis = 0))
```

```python
print( f'\n# 행 방향으로 평균 값을 구해주는 함수 applyByRow(row) 구현 \n{'-'*70}')

def applyByRow(row):
    mysum = 0
    for item in row:
        mysum += item
    return mysum / df.shape[1]

print(df.apply(applyByRow, axis = 1))
```
