# Series

```python
import pandas as pd

myindex = ['서울', '부산', '광주', '대구', '울산', '목포', '여수'] 
#변수(list 자료형 사용)

mylist = [50, 60, 40, 80, 70, 30, 20]
myseries = pd.Series(data=mylist, index=myindex)
print(myseries)
```

```python
print (' \n2번쨰 항목의 값 변경 ')
myseries.iloc[2] = 22
print(myseries)
```

```python
print(' \n2번째 부터 4번째 까지 항목의 값 변경 ')
myseries.iloc[2:5] = 33
print(myseries)
```

```python
print(' \n서울과 대구만 55로 변경 ')
myseries[['서울', '대구']] = 55
print(myseries)
```

```python
print(' \n짝수 행만 77로 변경')
myseries.iloc[0::2] = 77
```

```python
print(' \n시리즈 내용 확인 ')
print(myseries)
```

```python
print (' \n색인의 이름으로 값 읽기 ')
print(myseries[['대구']])
```

```python
print (' \n라벨의 이름으로 슬라이싱 ')
print(myseries['대구' : '목포'])
```

```python
print (' \n여러 개의 이름으로 데이터 읽기 ')
print(myseries[['대구', '여수']])
```

```python
print (' \n정수를 이용한 데이터 읽기 ')
print(myseries.iloc[2])
```

```python
print (' \n0, 2, 4번째 데이터 읽기 ')
print(myseries.iloc[0:5:2])
```

```python
print (' \n1,3,5번째 데이터 읽기 ')
print(myseries.iloc[[1,3,5]])
```

```python
print (' \n슬라이싱 사용하기 ')
#from <= result < to 
print(myseries.iloc[3:6]) 
```
