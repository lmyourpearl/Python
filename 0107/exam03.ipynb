```python
import pandas as pd
```

```python
myindex = ['마포구', '용산구', '서대문구', '동대문구', '은평구', '구로구', '강서구']
mylist = [40, 80, 70, 50, 60, 30, 20]
myseries = pd.Series(data=mylist, index=myindex)
myseries
```

```python
print( f'\n# 은평구만 조회하기 \n{'-'*40}')
myseries[['은평구']]
```

```python
print( f'\n# 서대문구부터 구로구까지 조회하기 \n{'-'*40}')
myseries['서대문구' : '구로구']
```

```python
print( f'용산구와 동대문구만 조회하기 \n{'-'*40}')
print(myseries[['용산구','동대문구']])
```

```python
print( f'2번째 요소만 조회하기 \n{'-'*40}')
print(myseries.iloc[2])
```

```python
print( f'0,2,4번째 데이터 조회하기 \n{'-'*40}')
print(myseries.iloc[0:5:2])
```

```python
print( f'1,3,4번째 데이터 조회하기 \n{'-'*40}')
print(myseries.iloc[[1, 3, 4]])
```

```python
print( f'슬라이싱을 사용하여 2번째부터 4번째까지 조회하기 \n{'-'*40}')
print(myseries.iloc[2:5])
```

```python
print( f'2번째 항목의 값을 99로 변경하기 \n{'-'*40}')
myseries.iloc[2] = 99
print(myseries)
```

```python
print( f'2번째부터 4번째까지를 66으로 변경하기 \n{'-'*40}')
myseries.iloc[2:4] = 66
print(myseries)
```

```python
print( f'마포구와 강서구만 55으로 변경하기 \n{'-'*40}')
myseries[['마포구', '강서구']] = 55
print(myseries)
```

```python
print( f'짝수 행만 77로 변경하기 \n{'-'*40}')
myseries.iloc[0::2] = 77
print(myseries)
```
