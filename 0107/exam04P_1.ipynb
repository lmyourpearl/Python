```python
import pandas as pd
import numpy as np
```

```python
myindex = ['김구', '이봉창', '안중근', '윤봉길']
mycolumns = ['강남구', '은평구', '마포구', '용산구']
mylist = list( 10 * onedata for onedata in range(1,17))
myframe = pd.DataFrame( np.reshape(mylist, (4,4)), index=myindex, columns = mycolumns)
myframe
```

```python
print(f"\n# 1번째 행 데이터를 조회하기 \n{'-'*40}")
myframe.iloc[1]
```

```python
print(f"\n# 1번째와 3번째 행 데이터를 조회하기 \n{'-'*40}")
myframe.iloc[[1,3]]
```

```python
print(f"\n# 윤봉길 행만 조회하기 \n{'-'*40}")
myframe.loc['윤봉길']
```

```python
print(f"\n# 이봉창과 윤봉길 행 조회하기 \n{'-'*40}")
myframe.loc[['윤봉길','이봉창']]
```

```python
print(f"\n# 윤봉길 행의 은평구 데이터만 조회하기 \n{'-'*40}")
myframe.loc[['윤봉길'], ['은평구']]
```

```python
print(f"\n# 김구와 이봉창의 용산구와 은평구 데이터 조회하기 \n{'-'*40}")
myframe.loc[['김구','이봉창'], ['용산구','은평구']]
```

```python
print(f"\n# 은평구의 값이 100이하인 행들을 조회하기 \n{'-'*40}")
myframe.loc[myframe['은평구'] <= 100]
```

```python
print(f"\n# 은평구의 값이 100인 행들을 조회하기 \n{'-'*40}")
myframe.loc[myframe['은평구'] == 100]
```

```python
print(f"\n# 김구부터 안중근까지 용산구 데이터를 80으로 변경하기 \n{'-'*40}")
myframe.loc['김구' : '안중근',['용산구']] = 80
myframe
```
