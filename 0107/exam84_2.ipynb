```python
import numpy as np 
import pandas as pd
```

```python
filename = './data/memberInfo_exam.csv'
df = pd.read_csv(filename, index_col = '이름')
df
```

```python
def sungjuk(score):
    if score >= 90:
        return '수'
    elif score >= 80:
        return '우'
    elif score >= 70:
        return '미'
    elif score >= 60:
        return '양'
    else:
        return '가'

sq = df['중간'].apply(sungjuk)
print(sq)
```
